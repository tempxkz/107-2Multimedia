```python
import sys
import cv2
import numpy as np
from matplotlib import pyplot as plt
#from google.colab import drive
#drive.mount('/content/drive')

#主函數
if __name__ =="__main__":
    if len(sys.argv)>1:
        image = cv2.imread('me.jpg', cv2.IMREAD_GRAYSCALE)
    else:
        print("Usge:python imgToArray.py imageFile")
    plt.imshow(image,cmap='gray')#https://blog.csdn.net/lanchunhui/article/details/66972783
    plt.show()
```

```python
image2 = cv2.imread('me.jpg', cv2.IMREAD_COLOR)#default
# plt.imshow(image2)
plt.imshow(cv2.cvtColor(image2,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image3 = cv2.imread('me.jpg', cv2.IMREAD_UNCHANGED)#including alpha channel
plt.imshow(cv2.cvtColor(image3,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image4 = cv2.imread('me.jpg', 0)# -1 0 1 2 3 4....................
plt.imshow(image4)
plt.title('invert')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
image5 = cv2.imread('me.jpg')
gray = cv2.cvtColor(image5, cv2.COLOR_BGR2GRAY)
gray
```

```python
plt.imshow(gray,cmap='gray')
plt.title('gray')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
image6 = cv2.imread('me.jpg')
b,g,r = cv2.split(image6)#slow
img = cv2.merge((r,g,b))#slow
plt.imshow(img)
plt.title('color')
plt.xticks([])
plt.yticks([])
plt.show()
```
